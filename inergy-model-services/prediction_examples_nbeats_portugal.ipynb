```python
import pandas as pd
ts = pd.read_csv("uc2_datasets/Portugal.csv",
                     sep=None,
                     header=0,
                     index_col=0,
                     parse_dates=True,
                     engine='python')
ts = ts[ts.index >= pd.Timestamp("2022-01-01 00:00:00")]
series = ts.to_json()
```

```python
ts
```

```python
import requests
import pandas as pd
from requests_toolbelt.multipart.encoder import MultipartEncoder

input_example = MultipartEncoder(
    fields={"n": "240",
            "series": series,
           }
)

result = requests.post(
    "http://0.0.0.0:3090/predict_nbeats_portugal", data=input_example, headers={"Content-Type": input_example.content_type}
).text
```
